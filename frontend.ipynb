```python
# Info = [name_text.get(),address_text.get(),tel_text.get(),email_text.get(),location_text.get(),no_of_units_text.get(),
#         type_text.get(),model_no_text.get(),serial_no_text.get(),sensor_defects_text.get(),date_of_installation_text.get(),
#         no_of_visits_text.get(),faults_list_text.get(),warranty_text.get()]
# Function to view all customer details in Database
def view_command():
    list1.delete(0,END)
    for row in backend.view():
        list1.insert(END,row)
```

```python
# Function to search for a particular Customer and their details
def search_command():
    list1.delete(0,END)
    for row in backend.search(name_text.get(),address_text.get(),tel_text.get(),email_text.get(),location_text.get()):
        list1.insert(END,row)
```

```python
# Function to add information of customers
def add_command():
    backend.insert(name_text.get(),address_text.get(),tel_text.get(),email_text.get(),location_text.get())
    list1.delete(0,END)
    list1.insert(END,(name_text.get(),address_text.get(),tel_text.get(),email_text.get(),location_text.get()))
```

```python
# Function to delete one particular customer
def delete_command():
    backend.delete(selected_tuple[0])
```

```python
# Function to Update information of a customer
def update_command():
    backend.update(selected_tuple[0],name_text.get(),address_text.get(),tel_text.get(),email_text.get(),location_text.get())
```

```python
# importing tkinter 
from tkinter import * 
from tkinter.ttk import * 
from tkinter.messagebox import askyesno
from datetime import datetime as dt

window=Tk()
window.wm_title("Customer Database")

def makelabel(parent,lbl,r,c,d=True,fon=None,colspan=None):
    l=Label(parent,text=lbl,font=fon)
    if d==True:
        l.grid(row=r,column=c,columnspan=colspan,sticky=W)
    else:
        l.grid(row=r,column=c,columnspan=colspan)
    return l
```

```python
makelabel(window,"AdvanceTech INDIA Pvt. Ltd.",0,0,True,("Times New Roman",28,"bold"),2)

# Creating Combobox
combo = ttk.Combobox(window, width="20", values=('Customer','Data Logger','Maintenance'))
combo.current(0) #set the selected item
combo.grid(column=2,row=0)#,sticky=E)

makelabel(window,"\nChoose any one in the given list to view / edit ↑ \n",2,0,False,("Arial Bold",16),4)

# To display live time on window
time1 = dt.now()
a=makelabel(window,time1,0,4,False,("Arial Bold",16))
def tick():
    global time1
    time2 = dt.now().strftime("%d-%m-%Y %H:%M:%S")
    if time2 != time1:
        time1 = time2
        a.config(text=time2)
    a.after(200, tick)
tick()

def closing_window():
    answer = askyesno("Exit","You want to exit the program?")
    if answer == True:
        window.destroy()
        
def makebutton(parent,text,r,c,w,comm,west=False):
    if west==True:
        l=Button(parent,text=text,width=w,command=comm).grid(row=r,column=c,sticky=W)
    else:
        l=Button(parent,text=text,width=w,command=comm).grid(row=r,column=c)

def makeentry(parent,text,r,c):
    l=Entry(parent,textvariable=text).grid(row=r,column=c,sticky=W)
    return l
    
def checkcmbo():
    a=combo.get().strip(" ")
    for label in window.grid_slaves():
        if 0 < int(label.grid_info()["row"]) < 5:
            label.grid_forget()
    if a == 'Customer':
        
        makelabel(window,"Name :",1,0,True)
        makelabel(window,"Address :",1,2,False)
        makelabel(window,"Mobile No :",2,0,True)
        makelabel(window,"Email Id :",2,2,False)
        makelabel(window,"Location :",3,0,True)

        name_text=StringVar()
        e1=makeentry(window,name_text,1,1)

        address_text=StringVar()
        e2=makeentry(window,address_text,1,3)

        tel_text=StringVar()
        e3=makeentry(window,tel_text,2,1)

        email_text=StringVar()
        e4=makeentry(window,email_text,2,3)

        location_text = StringVar()
        e5=makeentry(window,location_text,3,1)

    elif a == 'Data Logger':
        
        makelabel(window,"Name :",1,0,True)
        makelabel(window,"No.of Units :",1,2,False)
        makelabel(window,"Type :",2,0,True)
        makelabel(window,"Model.No :",2,2,False)
        makelabel(window,"Serial.No :",3,0,True)
        makelabel(window,"Sensor Defects :",3,2,False)
        makelabel(window,"Date of Installation :",4,0,True)
        
        name_text=StringVar()
        e6=makeentry(window,name_text,1,1)

        no_of_units_text=StringVar()
        e7=makeentry(window,no_of_units_text,1,3)

        type_text=StringVar()
        e8=makeentry(window,type_text,2,1)

        model_no_text=StringVar()
        e9=makeentry(window,model_no_text,2,3)

        serial_no_text=StringVar()
        e10=makeentry(window,serial_no_text,3,1)

        sensor_defects_text = StringVar()
        e11=makeentry(window,sensor_defects_text,3,3)

        date_of_installation_text = StringVar()
        e12=makeentry(window,date_of_installation_text,4,1)
        
    elif a == 'Maintenance':
        
        makelabel(window,"Name :",1,0,True)
        makelabel(window,"No. of Visits Done :",1,2,False)
        makelabel(window,"Faults List :",2,0,True)
        makelabel(window,"Warranty :",2,2,False)

        name_text=StringVar()
        e13=makeentry(window,name_text,1,1)

        no_of_visits_text=StringVar()
        e14=makeentry(window,no_of_visits_text,1,3)

        faults_list_text=StringVar()
        e15=makeentry(window,faults_list_text,2,1)

        warranty_text=StringVar()
        e16=makeentry(window,warranty_text,2,3)
    else:
        makelabel(window,"\nPlease,Choose any one in the given list ↑\n",2,0,False,("Arial Bold",18),4)
#         l1=Label(window,text="\nPlease,Choose any one in the given list ↑\n",font = ("Arial Bold",16))
#         l1.grid(row=2,column=0,columnspan=4)

makebutton(window,"Enter",0,3,5,checkcmbo,True)
# btn = Button(window, text="Enter",width=5,command=checkcmbo)
# btn.grid(row=0,column=3,sticky=W)

list1=Listbox(window,height=25,width=120)
list1.grid(row=7,column=0,rowspan=9,columnspan=5,sticky=W+E)

sb1=Scrollbar(window)
sb1.grid(row=7,column=4,rowspan=10,ipady=180,sticky=E)

sb2=Scrollbar(window,orient='horizontal')
sb2.grid(row=17,column=0,columnspan=5,sticky=N+S+W+E)

list1.configure(yscrollcommand=sb1.set)
sb1.configure(command=list1.yview)

list1.configure(xscrollcommand=sb2.set)
sb2.configure(command=list1.xview)

list1.bind('<<ListboxSelect>>',get_selected_row)

makebutton(window,"View all",5,0,14,view_command)
makebutton(window,"Search Entry",5,1,14,search_command)
makebutton(window,"Add Entry",5,2,14,add_command)
makebutton(window,"Update Selected",5,3,14,update_command)
makebutton(window,"Delete Entry",5,4,14,delete_command)
makebutton(window,"Close",18,4,14,closing_window)

makelabel(window,"Details",6,0,True,("Arial Bold",18,"bold"),6)
# lbl2 = Label(window, text="Details",font=("Arial Bold",18,"bold"))
# lbl2.grid(row=6,column=0,columnspan=6,sticky=W)

# all_widgets = window.winfo_children()
# for i in all_widgets:
#     window.columnconfigure(i,weight=1)

for i in range(0,4):
    window.grid_columnconfigure(i,weight=1)
    window.grid_rowconfigure(i,weight=2)

# Cpr =  Label(window,text="CopyRight@2020 Atech reserved",font=("Times New Roman",10))
# Cpr.grid(row=19,column=3)#,sticky=E)

window.mainloop()
```
